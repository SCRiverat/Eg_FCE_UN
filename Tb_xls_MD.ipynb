```
import pandas as pd
from pyxlsb import open_workbook as open_xlsb

def excel_to_md(excel_file, sheet_name):
    try:
        # Leer el archivo de Excel utilizando pyxlsb
        with open_xlsb(excel_file) as wb:
            with wb.get_sheet(sheet_name) as sheet:
                # Leer los datos de la hoja y almacenarlos en una lista de filas
                rows = []
                for row in sheet.rows():
                    rows.append([item.v for item in row])

        # Crear un DataFrame de pandas con los datos
        df = pd.DataFrame(rows)

        # Convertir el DataFrame en una tabla en formato Markdown
        table = df.to_markdown(index=False)

        return table
    except ImportError:
        print("El paquete 'pyxlsb' no está instalado. Instala 'pyxlsb' utilizando pip.")
        return None

# Ejemplo de uso
archivo_excel = 'Datos.xlsb'  # Reemplaza con la ruta de tu archivo Excel
hoja_excel = 'Hoja1'  # Reemplaza con el nombre de la hoja en tu archivo Excel
tabla_md = excel_to_md(archivo_excel, hoja_excel)

if tabla_md is not None:
    print(tabla_md)
```

```
archivo_md = 'tabla.md'  # Ruta donde se guardará el archivo .md
with open(archivo_md, 'w') as file:
    file.write(tabla_md)
```
